```python
# Allows plotting
%matplotlib inline 
# Import libraries and functions
from imageClassification import *
# Plotting
import seaborn as sns; sns.set() 
from keras import regularizers
```

```python
dataset_path = "animals/"
model_path = "output/interesting/"
label_path = "output/interesting/"
plot_path = "output/interesting/"
report_path = "output/interesting/"
```

```python
dataset, labels = loadImgDataset(dataset_path)
```

```python
name = "default"
trainX, testX, trainY, testY, lb = imagePreprocesingFFNN(
    dataset,
    labels, 
    height=32, 
    width=32, 
    test_size=0.25, 
    random_state=42, 
    PCA=False,
    scaled=False,
    minmax=False,
    RGB=True,
    k=0,
    lb_conf=1, 
    augmented=False, 
    rot=False, 
    flip=False,
    sd=0)
```

```python
default = train_clasificationNN( 
    trainX, testX, trainY, testY, lb,
    name=name,
    epochs=75, 
    batch_size=32,
    report_path=report_path,
    show=True,
    plot_path=plot_path
)
```

```python
saveModel(name, default, lb, model_path=model_path, label_path=label_path)
```

```python
name = "scale"
trainX_scale, testX_scale, trainY_scale, testY_scale, lb_scale = imagePreprocesingFFNN( 
    dataset, 
    labels,
    height=32,
    width=32,
    test_size=0.25, 
    random_state=42, 
    PCA=False,
    scaled=True,
    minmax=False,
    RGB=False,
    lb_conf=1, 
    augmented=True, 
    rot=False,
    flip=False,
    sd=0)
```

```python
scale = train_clasificationNN( 
    trainX_scale, testX_scale, trainY_scale, testY_scale, lb_scale, 
    name=name,
    model=Sequential([
                Dropout(0.2),
                Dense(16, input_shape=(256,), kernel_regularizer=regularizers.l2(0.1)),
                Activation("relu"),
                Dropout(0.5),
                Dense(8, kernel_regularizer=regularizers.l2(0.1)),
                Activation("relu"),           
                Dropout(0.5),
                Dense(3, kernel_regularizer=regularizers.l2(0.1)),
                Activation("relu"),                   
                Dense(3, kernel_regularizer=regularizers.l2(0.1)),
                Activation("relu"),                   
                Dense(3),
                Activation('softmax')
            ]),
    epochs=50, 
    batch_size=32,
    report_path=report_path,
    show=True,
    plot_path=plot_path
)
```

```python
saveModel(name, scale, lb_scale, model_path=model_path, label_path=label_path)
```

```python
name = "PCA"
trainX_PCA, testX_PCA, trainY_PCA, testY_PCA, lb_PCA = imagePreprocesingFFNN( 
    dataset, 
    labels,
    height=64,
    width=64,
    test_size=0.25, 
    random_state=42, 
    PCA=True,
    k=256,
    scaled=False,
    minmax=False,
    RGB=False,
    lb_conf=1, 
    augmented=True, 
    rot=False,
    flip=True,
    sd=0)
```

```python
PCA_nesterov3_model = Sequential([
                Dropout(0.1),
                Dense(32, input_shape=(256,), kernel_regularizer=regularizers.l2(0.15)),
                Activation("relu"),
                Dropout(0.3),
                Dense(16, kernel_regularizer=regularizers.l2(0.15)),
                Activation("relu"),           
                Dense(3),
                Activation('softmax')
            ])
```

```python
name = "PCA_nesterov3"
PCA_nesterov3 = train_clasificationNN( 
    trainX_PCA, testX_PCA, trainY_PCA, testY_PCA, lb_PCA, 
    name=name,
    model=PCA_nesterov3_model,
    opt=SGD(lr=0.01, momentum=.3, nesterov=True),
    epochs=50, 
    batch_size=32,
    report_path=report_path,
    show=True,
    plot_path=plot_path
)
```

```python
saveModel(name+"flip", PCA_nesterov3, lb_PCA, model_path=model_path, label_path=label_path)
```
